## Conversão de imagem em um Esboço a Lápis usando Python

Isso vai ser interessante. Estaremos escrevendo o código passo a passo com a explicação.

Usaremos a biblioteca OpenCV para este projeto. Instale-o usando o comando pip install opencv-python .

Etapa 1: encontre uma imagem que você deseja converter em um esboço a lápis.

Vamos usar uma imagem de cachorro. Você pode escolher o que quiser.

Passo 2 : Leia a imagem no formato RBG e depois converta-a para uma imagem em tons de cinza. Agora, a imagem é transformada em uma foto clássica em preto e branco.

```python
from IPython.display import Image
Image('D:/raiz/VIDA/foto_cobra_Python.jpg')
```

```python
# Importando a Biblioteca
import cv2

#Lendo a Imagem
image = cv2.imread("D:/raiz/VIDA/foto_cobra_Python.jpg")
```

```python
#Covertendo BGR imagem em Escala de cinza
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

Passo 3: Inverta a imagem em tons de cinza também chamada de imagem negativa, esta será nossa imagem em tons de cinza invertida. A inversão é basicamente usada para aprimorar detalhes.

```python
#Inversão de Imagem

inverted_image = 255 - gray_image
```

Passo 4: Por fim, crie o esboço a lápis misturando a imagem em tons de cinza com a imagem borrada invertida. Isso é feito dividindo a imagem em tons de cinza pela imagem borrada invertida.

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

Agora temos nosso pencil_sketch. Então, exiba-o usando o OpenCV

```python
from IPython.display import Image
Image('D:/raiz/VIDA/Esboço_Python.png')# Esta é a imagem que você verá ao executar o código
```

```python
cv2.imshow("Original Image", image)
cv2.imshow("Pencil Sketch of foto", pencil_sketch)
cv2.waitKey(0)
```
